**Basic Number Operations**

Arithmetic Operators

```python
x = 10
y = 3
print(x + y)   # Output: 13
print(x - y)   # Output: 7
print(x * y)   # Output: 30
print(x / y)   # Output: 3.3333333333333335
print(x // y)  # Output: 3
print(x % y)   # Output: 1
print(x ** y)  # Output: 1000

fname = "Sakda"
lname = "Homhuan"

print(fname+" "+lname)
```
