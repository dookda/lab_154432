การใช้ Loop แบบต่างๆ

1. **การใช้ `for` Loop**

```python
# Iterating over a list
fruits = ['apple', 'banana', 'cherry']
for f in fruits:
    print(f)
```

```python
# Using the `range()` function
for i in range(15):  # Will iterate over 0, 1, 2, 3, 4
    print(i)
```

```python
# Iterating over a dictionary
person = {'name': 'Alice', 'age': 25}
# print(person.items())
for key, value in person.items():
    print(key, value)
```

2. **การใช้ `while` Loop**

```python
# Simple `while` loop
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Using a `while` loop with `break`
count = 0
while True:  # This creates an infinite loop
    if count >= 5:
        break  # Break out of the loop if condition is met
    print(count)
    count += 1
```

```python
# Using a `while` loop with `continue`
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue  # Skip the rest of the loop for even numbers
    print(count)  # This will print only odd numbers
```

3. **การใช้ Loop Controls**

```python
# `else` with `for` loop
for i in range(5):
    print(i)
else:
    print("Done!")  
```

```python
# `else` with `while` loop
count = 0
while count < 5:
    print(count)
    count += 1
else:
    print("Count is no longer less than 5")
```

4. **อื่นๆ**

```python
# loop in List
[print(i) for i in range(5)]
```
